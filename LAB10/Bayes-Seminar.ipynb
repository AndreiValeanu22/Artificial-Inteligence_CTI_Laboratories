## Explicații — probabilități de bază

<div style="background: linear-gradient(90deg, #fff7ed, #fdf4ff); border-left: 6px solid #f97316; padding: 10px 14px; margin: 12px 0; font-family: system-ui, sans-serif; color: #1e1b4b;">

**Pe scurt (începători — rețele Bayesiene):**

- **Variabilă aleatoare**: rezultat incert înainte de observație.
- **p(X)**: probabilitatea evenimentului X (între 0 și 1).
- **Independență**: X ⊥ Y ⇒ p(X∩Y) = p(X)p(Y).
- **Condiționare p(X|Y)**: probabilitatea lui X după ce știm Y.


</div>


## Explicații — Teorema lui Bayes

<div style="background: linear-gradient(90deg, #fff7ed, #fdf4ff); border-left: 6px solid #f97316; padding: 10px 14px; margin: 12px 0; font-family: system-ui, sans-serif; color: #1e1b4b;">

**Pe scurt (începători — rețele Bayesiene):**

- **Rol**: din observație (efect) inferăm cauza.
- **Posterior** p(cauză|efect) ∝ verosimilitate × prior (normalizat).
- **Numitor**: probabilitatea totală a observației; uneori se evită prin rapoarte (artificiul din seminar).
- **Exemplul gripei din material:** la formula (10) numitorul corect este suma de produse p(+|Gripa)p(Gripa) + p(+|¬Gripa)p(¬Gripa), nu expresia cu `0.9 + 0.01 + …` din tipărire (eroare de formă).


</div>

<div style="background: linear-gradient(90deg, #fff7ed, #fdf4ff); border-left: 6px solid #f97316; padding: 10px 14px; margin: 12px 0; font-family: system-ui, sans-serif; color: #1e1b4b;">

**Pe scurt (începători — rețele Bayesiene):**

- **Nod** = variabilă; **arc** = influență directă (de obicei cauzală).
- **CPT**: probabilitățile nodului condiționate de părinți.
- **Economie de parametri** față de tabelul joint complet.

## Explicații — Independență condițională și d-separabilitate

<div style="background: linear-gradient(90deg, #fff7ed, #fdf4ff); border-left: 6px solid #f97316; padding: 10px 14px; margin: 12px 0; font-family: system-ui, sans-serif; color: #1e1b4b;">

**Pe scurt (începători — rețele Bayesiene):**

- **X ⊥ Z | Y**: ținând cont de Y, Z nu modifică credința despre X.
- **d-separare**: mulțimea observată blochează toate căile neorientate ⇒ independență condițională.
- **Colider** (A→C←B): blocat dacă C și descendenții nu sunt observați; se deschide la observare pe C sau descendent.
